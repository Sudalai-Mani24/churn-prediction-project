```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv("processed_ott_data.csv")
```

```python
from sklearn.preprocessing import StandardScaler

# Select relevant numerical features for clustering
features = ['age', 'weekly_mins_watched', 'videos_watched', 'customer_support_calls', 'no_of_days_subscribed']

# Standardizing the features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(df[features])
```

```python
#Determine the Optimal Number of Clusters (Elbow Method)
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Finding the optimal number of clusters using the Elbow Method
inertia = []
K_range = range(1, 11)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(scaled_features)
    inertia.append(kmeans.inertia_)

# Plot the Elbow curve
plt.figure(figsize=(8, 5))
plt.plot(K_range, inertia, marker='o', linestyle='--')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal K')
plt.show()
```

```python
# Apply K-Means with the chosen number of clusters (e.g., K=3)
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
df['customer_segment'] = kmeans.fit_predict(scaled_features)
```

```python
# Understanding the characteristics of each cluster
segment_analysis = df.groupby('customer_segment')[features].mean()
print(segment_analysis)
```

```python
print(df.columns)

```

```python
#Visualize the Clusters
#Plot clusters using PCA (Principal Component Analysis) to reduce dimensions.
import seaborn as sns
from sklearn.decomposition import PCA

# Reduce to 2D for visualization
pca = PCA(n_components=2)
pca_features = pca.fit_transform(scaled_features)

# Add PCA results to DataFrame
df['pca1'] = pca_features[:, 0]
df['pca2'] = pca_features[:, 1]

# Scatter plot
plt.figure(figsize=(8,6))
sns.scatterplot(x=df['pca1'], y=df['pca2'], hue=df['customer_segment'], palette='viridis')
plt.title('Customer Segmentation (K-Means Clustering)')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.show()
```
